```python
import numpy as np
import pandas as pd

from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
```

```python
data= pd.read_csv("SalaryData_Train (1).csv")
test=pd.read_csv("SalaryData_Test (1).csv")
```

```python
df1 = data.copy()
df2 = test.copy()
```

```python
df1.isnull().sum()
df2.isnull().sum()
```

#### one hot encoding

```python
train1 = df1.iloc[:,0:13]

train1 = pd.get_dummies(train1)
train1
```

```python
test1 = test.iloc[:,0:13]

test1 = pd.get_dummies(test1)
test1
```

```python
ftrain = pd.concat([train1, df1['Salary']],axis=1)
ftrain
```

```python
ftest = pd.concat([test1, df2['Salary']],axis=1)
ftest
```

```python
# Finaltrain data:
X= ftrain.values[:,0:102]
Y= ftrain.values[:,102]
```

```python
#Finaltest data:
x= ftest.values[:,0:102]
y= ftest.values[:,102]
```

### Naive Bayes Model
Multinomial Naive Bayes

```python
classifier_mb = MB()
classifier_mb.fit(X,Y)
train_pred_m = classifier_mb.predict(X)
accuracy_train_m = np.mean(train_pred_m==Y)
```

```python
test_pred_m = classifier_mb.predict(x)
accuracy_test_m = np.mean(test_pred_m==y)
```

```python
print('Training accuracy is:',accuracy_train_m,'\n','Testing accuracy is:',accuracy_test_m)
```

###Gaussian Naive Bayes 

```python
classifier_gb = GB()
classifier_gb.fit(X,Y) 
# we need to convert tfidf into array format which is compatible for gaussian naive bayes
train_pred_g = classifier_gb.predict(X)
accuracy_train_g = np.mean(train_pred_g==Y)
```

```python
test_pred_g = classifier_gb.predict(x)
accuracy_test_g = np.mean(test_pred_g==y)
```
