```python
airlines=airlines.drop(['ID#','Award?'],axis=1)
airlines
```

```python
from sklearn.preprocessing import MinMaxScaler
transformation=MinMaxScaler()
airlines1=pd.DataFrame(transformation.fit_transform(airlines.iloc[:,0:]))
airlines1
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
a=linkage(airlines1,method='average',metric='euclidean')
plt.figure(figsize=(15,5))
plt.title("Hierachinal clustering Dendrogram")
plt.xlabel("index")
plt.ylabel("distance")
sch.dendrogram(a)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
H_complete=AgglomerativeClustering(n_clusters=4, linkage="average",affinity="euclidean").fit(airlines1)
cluster_labels=pd.Series(H_complete.labels_)
cluster_labels
airlines['clust']=cluster_labels
airlines
```

```python
airlines.iloc[:,1:].groupby(airlines.clust).mean()
```

```python
airlines2=airlines[(airlines.clust==0)]
airlines2
```

```python
airlines3=airlines[(airlines.clust==1)]
airlines3
```

```python
import sklearn
sklearn.metrics.silhouette_score(airlines1,H_complete.labels_)
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(airlines1)
    WCSS.append(clf.inertia_) 
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(airlines1)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python
neighbors=NearestNeighbors(n_neighbors=8)
nbrs=neighbors.fit(airlines1)
distances, indices=nbrs.kneighbors(airlines1)
```

```python
distances=np.sort(distances,axis=0)
distances=distances[:,1]
plt.plot(distances)
```

```python
dbscan=DBSCAN(eps=0.475, min_samples=5)
dbscan.fit(airlines1)
```

```python
dbscan.labels_
```

```python
c1=pd.DataFrame(dbscan.labels_,columns=['clusters'])
```
